```python
#-*- coding:utf-8 -*-
 
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import miracle
import time
import os
import tableauUtils
 
miracle = miracle.miracle()
util = tableauUtils.tableauUtils()
 
download_path = '/Users/ace-devnew1/Desktop/python/tableau/temp/'
driver_path = '/Users/ace-devnew1/Desktop/python/tableau/geckodriver'
url = 'https://us-west-2b.online.tableau.com/#/site/acea/views/_13/sheet17'
channel = 'cpm_game-design'
tableau_id = 'ace@aceproject.co.kr'

#패스워드 입력 필요
password = 'pw'
 
def download_image(filename):
    time.sleep(5)

    driver.find_element_by_css_selector("#download-ToolbarButton").click() 
    driver.find_element_by_xpath('//*[@id="DownloadDialog-Dialog-Body-Id"]/div/button[1]').click()
 
    while True:
        if not os.path.exists(download_path + '통합뷰.png'):
            time.sleep(3)
        else:
            break
 
    os.rename(download_path + '통합뷰.png', download_path + filename + '.png')

def tab_click(tabNo):
    time.sleep(5) 
    driver.find_element_by_xpath("//span[contains(@class,'tabStoryPoint')][" + str(tabNo) + "]").click()
    
successCount = 0
 
while successCount < 5 :
    successCount = 0
    try:
        options = Options()
 
        options.set_preference('browser.download.folderList', 2) # custom location
        options.set_preference('browser.download.manager.showWhenStarting', False)
        options.set_preference('browser.download.dir', download_path)
        options.set_preference("browser.download.manager.alertOnEXEOpen", False)
        options.set_preference("browser.download.manager.closeWhenDone", False)
        options.set_preference("browser.download.manager.focusWhenStarting", False)
        options.set_preference("browser.helperApps.neverAsk.saveToDisk", "image/png")
        options.headless = True
        driver = webdriver.Firefox(options=options,executable_path=driver_path)
        driver.implicitly_wait(10)
        
        driver.get(url) 
        driver.find_element_by_id('email').send_keys(tableau_id)
        driver.find_element_by_id('password').send_keys(password)
        driver.find_element_by_id('login-submit').click()
        driver.switch_to.frame(driver.find_element_by_css_selector("#viz > iframe"))
        tab_click(1)
        
        download_image('DAILY REPORT')
        successCount += 1
        tab_click(2)
 
        download_image('AP')
        successCount += 1
        tab_click(3)
        
        download_image('CASH')
        successCount += 1
        tab_click(4)
 
        download_image('SALES')
        successCount += 1
        tab_click(5)
 
        download_image('NU') 
        successCount += 1
 
        miracle.send_file(channel, download_path + 'DAILY REPORT.png')
        miracle.send_file(channel, download_path + 'AP.png')
        miracle.send_file(channel, download_path + 'CASH.png')
        miracle.send_file(channel, download_path + 'SALES.png')
        miracle.send_file(channel, download_path + 'NU.png')
        miracle.send_message(channel, '자세한 사항은 태블로 페이지(' + url + ')에서 확인해주세요!')

    except Exception as e: 
        print(e)
        pass
 
    finally:
        driver.quit()
        util.removeAllFile(download_path)

Collapse
```
