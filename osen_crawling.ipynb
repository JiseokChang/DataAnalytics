```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import urllib.request
import os
import re
  
#검색 선수명 입력
player = "선수명 입력"
url ="http://photoout.osen.co.kr/photoout2/login2"
  
#검색 시간 조건 입력
fromDate = '날짜 입력'
  
#검색 구분 조건(110 = 야구)
code = '110'
os.makedirs(player)
  
#로그인: id, pw 입력
driver = webdriver.Chrome(executable_path = 'C:\\Users\\GD-gorgo10\\Desktop\\chromedriver_win32\\chromedriver.exe')
driver.get(url)
driver.implicitly_wait(3)
driver.maximize_window()
driver.implicitly_wait(1)
driver.find_element_by_id('id').send_keys("id")
driver.find_element_by_id('password').send_keys("pw")
select = driver.find_element_by_xpath('/html/body/div/form/p[3]/input')
select.click()
driver.implicitly_wait(3)
  
#30offset 단위 페이지 이동
for i in range(201):
    offset = i*30
    new_url = "http://photoout.osen.co.kr/photoout2/photolist2/?keyword="+player+"&offset="+str(offset)+"&fromDate="+fromDate+"&code="+code
    driver.get(new_url)
    driver.implicitly_wait(1)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
     
    img = soup.find_all("img")
    detail = soup.find_all("button")
    count = 1
     
    #페이지별 img_src 및 detail 저장
    for a in img:
        img_src = a.get("src")
        detail_page = detail[count].get("onclick")
        split = detail_page.split("'")
        detail_url = "http://photoout.osen.co.kr" + split[1]
        driver.get(detail_url)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        td = soup.find_all("td")
        title= td[1].string
        title = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\'|\(\)\[\]\<\>`\'...》]', '', title)
        upload_date = td[7].string
         
        file_name = title + "_" + upload_date + ".jpg"
        urllib.request.urlretrieve(img_src, "C:\\Users\\GD-gorgo10\\Desktop\\"+player+"\\"+file_name)
        count = count+2
        driver.implicitly_wait(3)
         
print(player+" "+"다운로드가 완료되었습니다.")
```
